```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path="/content/atenciones_abogadas_de_las_mujeres_2023_1006_2.csv"

# Crear un DataFrame de ejemplo
df = pd.read_csv(path)

# Calcular value_counts para la columna A
#conteo = df['anio'].value_counts()

# Obtener el conteo de cada categoría
count = df["anio"].value_counts()

# Crear el diagrama de dispersión


#sns.scatterplot(x=count.index, y=count)

sns.lineplot(x=count.index, y=count)

# Mostrar el gráfico
plt.show()
```
